```python
# Label encode: Gender column 
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
df['Gender'] = encoder.fit_transform(df['Gender'])
```

```python
X = df.drop('Purchased', axis = 1)
Y = df['Purchased']

X
```

```python
Y
```

```python
# splitting the data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)


# SCALING
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()
X_train = ss.fit_transform(X_train) #z-score se apan scale kr rahe
X_test = ss.transform(X_test) #mean and std X_train ka hai, apan sirf transform kr rahe
```

```python
# X_train.shape
X_train
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, Y_train) #model ban jata fit krne se
```

```python
Y_pred = model.predict(X_test)
```

```python
Y_pred
Y_train.value_counts()
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(Y_test, Y_pred)
cm
```

```python
# Performance metrics
tp = cm[0][0]
fp = cm[0][1]
fn = cm[1][0]
tn = cm[1][1]
```

```python
# Precision:
precision = tp/(tp+fp)

# Recall:
recall = tp/(tp+fn)

# Accuracy:
accuracy = (tp+tn)/(tp+fp+fn+tn)

# f1 score:
f1_score = 2 * (recall*precision)/(recall + precision)

print(precision, recall, accuracy, f1_score)
```
