```python
# Replace missing values with mean, median and mode

# 1. Replacing with mean in math score col
math_score_mean = df['math score'].mean()
df['math score'] = df['math score'].fillna(math_score_mean)
```

```python
df.isnull().sum()
```

```python
reading_score_mode = df['reading score'].mode()[0]
print(reading_score_mode.tolist())
df['reading score'].fillna(reading_score_mode, inplace=True)
```

```python
df.isnull().sum()
```

```python
df.head()
```

```python
# replacing with median
writing_score_median = df['writing score'].median()
df['writing score'].fillna(writing_score_median, inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

# Min max normalization

```python
# Perform min max normalization on 'math score' column:
min_ms = df['math score'].min()
max_ms = df['math score'].max()

df['normalized math scores'] = [( (score - min_ms)/(max_ms - min_ms)) for score in df['math score']]

# drop math score column
df.drop(['math score'], axis = 1, inplace=True)
```

```python
df.head()
```

# Z score normalization

```python
# Z score normalization on reading score
mean_rs = df['reading score'].mean()
std_rs = df['reading score'].std()

df['reading score'] = [((score - mean_rs)/std_rs) for score in df['reading score']]
```

```python
df.head()
```

# Decimal Scaling

```python
# Decimal scaling on writing score
max_rs = df['writing score'].max()
max_rs
```

```python
df['writing score'] = [(score/100) for score in df['writing score']]
```
